```python
# ==========================
# Updated H2O AutoML Script
# Prints best model + hyperparameters with high precision and prediction output
# ==========================

!pip install -f http://h2o-release.s3.amazonaws.com/h2o/latest_stable_Py.html h2o

import h2o
from h2o.automl import H2OAutoML
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

h2o.init()

# Load data
df = pd.read_csv("https://raw.githubusercontent.com/apownukepcc/spring-2025-datathon/main/009-Dataset-For-Predictions-With-Specific-Emissions.csv")
df['date'] = pd.to_datetime(df['date'])
specific_test_date = pd.to_datetime("2022-07-15")
specific_test_row = df[df['date'] == specific_test_date]
df = df[df['date'] != specific_test_date]

feature_cols = ['tavg', 'tmin', 'tmax', 'prcp', 'snow', 'wdir', 'wspd', 'pres']
target_cols = [
    'SO2TONS_per_LOADMWBA', 'SO2TONS_per_LOADMWBT',
    'NH3TONS_per_LOADMWBA', 'NH3TONS_per_LOADMWBT',
    'NOXTONS_per_LOADMWBA', 'NOXTONS_per_LOADMWBT',
    'COTONS_per_LOADMWBA', 'COTONS_per_LOADMWBT']

df = df.dropna(subset=feature_cols + target_cols)

results = []
comparison_data = {}

for target in target_cols:
    print(f"\n‚è≥ Training AutoML model for: {target}")

    h2o_df = h2o.H2OFrame(df[feature_cols + [target]])
    train, test = h2o_df.split_frame(ratios=[0.8], seed=1)

    aml = H2OAutoML(max_runtime_secs=120, seed=1)
    aml.train(x=feature_cols, y=target, training_frame=train)

    test_row = h2o.H2OFrame(specific_test_row[feature_cols])
    prediction = None
    best_model = aml.leader

    try:
        prediction = best_model.predict(test_row).as_data_frame().iloc[0, 0]
        print(f"‚úÖ Used model: {best_model.model_id} ({best_model.algo})")
        print(f"   Prediction for {specific_test_date.date()}: {prediction:.10f}")
    except:
        for model_id in aml.leaderboard.as_data_frame()['model_id']:
            model = h2o.get_model(model_id)
            if "StackedEnsemble" not in model.algo:
                try:
                    prediction = model.predict(test_row).as_data_frame().iloc[0, 0]
                    best_model = model
                    print(f"‚úÖ Fallback model: {best_model.model_id} ({best_model.algo})")
                    print(f"   Prediction for {specific_test_date.date()}: {prediction:.10f}")
                    break
                except:
                    continue

    actual = specific_test_row[target].values[0] if prediction is not None else None
    if actual is not None:
        print(f"   Actual Value: {actual:.10f}")

    comparison_data[target] = {'Actual': actual, 'Predicted': prediction}

    try:
        perf = best_model.model_performance(test_data=test)
        r2 = perf.r2()
        rmse = perf.rmse()
        mae = perf.mae()

        print(f"   R2: {r2:.10f}, RMSE: {rmse:.10f}, MAE: {mae:.10f}")

        try:
            params = best_model.params
            param_summary = {k: v['actual'] for k, v in params.items() if 'actual' in v}
            print("   üîß Best Hyperparameters:")
            for key, val in param_summary.items():
                print(f"      {key}: {val}")
        except:
            print("   ‚ö†Ô∏è Unable to extract hyperparameters.")

        results.append({
            'Parameter': target,
            'Model': best_model.algo,
            'R2': r2,
            'RMSE': rmse,
            'MAE': mae,
            'Model ID': best_model.model_id
        })

    except:
        print(f"‚ùå Model evaluation failed for {target}")
        results.append({
            'Parameter': target,
            'Model': 'Failed',
            'R2': None,
            'RMSE': None,
            'MAE': None,
            'Model ID': 'N/A'
        })

# Summary tables
results_df = pd.DataFrame(results)
comparison_df = pd.DataFrame(comparison_data).T

print("\nüìä AutoML Accuracy Metrics:")
print(results_df.to_string(float_format='%.10f'))

print(f"\nüìÖ Actual vs Predicted for {specific_test_date.date()}:")
print(comparison_df.applymap(lambda x: f"{x:.10f}" if isinstance(x, (float, int)) else x))

# Plot
plot_df = comparison_df.dropna()
fig, ax = plt.subplots(figsize=(12, 6))
x = np.arange(len(plot_df))
width = 0.35
bars1 = ax.bar(x - width/2, plot_df['Actual'], width, label='Actual')
bars2 = ax.bar(x + width/2, plot_df['Predicted'], width, label='Predicted')
ax.set_ylabel('Emissions per Load (Tons/MW)')
ax.set_title(f'Actual vs Predicted Emissions for {specific_test_date.date()}')
ax.set_xticks(x)
ax.set_xticklabels(plot_df.index, rotation=45, ha='right')
ax.legend()

def add_labels(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.10f}',
                    xy=(bar.get_x() + bar.get_width()/2, height),
                    xytext=(0, 3), textcoords="offset points",
                    ha='center', va='bottom')

add_labels(bars1)
add_labels(bars2)
plt.tight_layout()
plt.grid(True, axis='y', linestyle='--', alpha=0.6)
plt.show()
```
