```python
import pandas as pd
from google.colab import drive

# Mount Google Drive (if not already mounted)
drive.mount('/content/drive')

# URL of the cleaned dataset
url = "https://raw.githubusercontent.com/apownukepcc/spring-2025-datathon/main/007-Dataset-For-Predictions-Cleaned.csv"

# Load the dataset
df = pd.read_csv(url)

# List of emission parameters
emission_params = ['SO2TONS', 'NH3TONS', 'NOXTONS', 'COTONS']

# Calculate specific emissions
for param in emission_params:
    df[f'{param}_per_LOADMWBA'] = df[param] / df['LOADMWBA']
    df[f'{param}_per_LOADMWBT'] = df[param] / df['LOADMWBT']

# Define output file path in Google Drive
output_file = "/content/drive/My Drive/009-Dataset-For-Predictions-With-Specific-Emissions.csv"

# Save the updated dataset to Google Drive
df.to_csv(output_file, index=False)
print(f"Updated dataset saved to Google Drive: {output_file}")
```
