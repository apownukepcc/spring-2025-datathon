```python
# Install necessary packages (uncomment if needed)
!pip install catboost

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Conv1D, Flatten, LSTM, InputLayer
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# -------------------------------
# Load and Preprocess the Data
# -------------------------------
data_url = "https://raw.githubusercontent.com/apownukepcc/ForecastingDailyEmissions/refs/heads/main/SO2TONS_dataset.csv"
data = pd.read_csv(data_url)
data['date'] = pd.to_datetime(data['date'])
# Filter for peak season (May-August)
peak_season_months = [5, 6, 7, 8]
data = data[data['date'].dt.month.isin(peak_season_months)]
predictors = ['tavg', 'tmin', 'tmax', 'prcp', 'snow', 'wdir', 'wspd', 'pres']
target = 'Emissions_Load'
data = data.dropna(subset=predictors + [target])
input_dim = len(predictors)

# -------------------------------
# Define Function to Compute Metrics
# -------------------------------
def compute_metrics(y_true, y_pred):
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mae  = mean_absolute_error(y_true, y_pred)
    r2   = r2_score(y_true, y_pred)
    mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
    return rmse, mae, r2, mape

# -------------------------------
# Define Neural Network Configurations
# -------------------------------
# Each configuration is a dictionary entry containing:
#   - 'builder': a function that builds and compiles a model given a learning rate.
#   - 'epochs': number of training epochs.
#   - 'lr': learning rate.
# We define several architectures:
#   1. One hidden layer with 32 neurons.
#   2. Two hidden layers (64, 32) with ReLU.
#   3. Three hidden layers (128, 64, 32) with Tanh.
#   4. Two hidden layers (64, 32) with Dropout.
#   5. Two hidden layers (64, 32) with ReLU but longer training (more epochs).
#   6. Two hidden layers (64, 32) with ReLU but lower learning rate.
#
tf_nn_configs = {
    "MLP_1layer_32_E50_LR0.001": {
        "builder": lambda lr: Sequential([
            InputLayer(input_shape=(input_dim,)),
            Dense(32, activation='relu'),
            Dense(1)
        ]),
        "epochs": 50,
        "lr": 0.001
    },
    "MLP_2layer_64-32_E50_LR0.001": {
        "builder": lambda lr: Sequential([
            InputLayer(input_shape=(input_dim,)),
            Dense(64, activation='relu'),
            Dense(32, activation='relu'),
            Dense(1)
        ]),
        "epochs": 50,
        "lr": 0.001
    },
    "MLP_3layer_128-64-32_E50_LR0.001": {
        "builder": lambda lr: Sequential([
            InputLayer(input_shape=(input_dim,)),
            Dense(128, activation='tanh'),
            Dense(64, activation='tanh'),
            Dense(32, activation='tanh'),
            Dense(1)
        ]),
        "epochs": 50,
        "lr": 0.001
    },
    "MLP_2layer_64-32_Dropout_E50_LR0.001": {
        "builder": lambda lr: Sequential([
            InputLayer(input_shape=(input_dim,)),
            Dense(64, activation='relu'),
            Dropout(0.3),
            Dense(32, activation='relu'),
            Dropout(0.3),
            Dense(1)
        ]),
        "epochs": 50,
        "lr": 0.001
    },
    "MLP_2layer_64-32_E100_LR0.001": {
        "builder": lambda lr: Sequential([
            InputLayer(input_shape=(input_dim,)),
            Dense(64, activation='relu'),
            Dense(32, activation='relu'),
            Dense(1)
        ]),
        "epochs": 100,
        "lr": 0.001
    },
    "MLP_2layer_64-32_E50_LR0.0005": {
        "builder": lambda lr: Sequential([
            InputLayer(input_shape=(input_dim,)),
            Dense(64, activation='relu'),
            Dense(32, activation='relu'),
            Dense(1)
        ]),
        "epochs": 50,
        "lr": 0.0005
    }
}

# -------------------------------
# Compare Configurations for Each Powerplant
# -------------------------------
# We'll store summary results for each powerplant in a dictionary.
source_summaries = {}

for source in data['Source'].unique():
    print(f"\n=== Processing powerplant: {source} ===")
    data_source = data[data['Source'] == source].copy()

    # Skip powerplants with too few data points
    if data_source.shape[0] < 10:
        print(f"Not enough data for {source}, skipping...")
        continue

    # Prepare X and y for the current powerplant
    X_source = data_source[predictors].values
    y_source = data_source[target].values.reshape(-1, 1)

    # Split data
    X_train, X_test, y_train, y_test = train_test_split(X_source, y_source, test_size=0.2, random_state=42)

    # Dictionary to collect performance metrics for each configuration
    perf_metrics = {}

    # Loop over each neural network configuration
    for config_name, config in tf_nn_configs.items():
        print(f"Training config {config_name} for {source}")
        # Build the model using the provided builder lambda
        model = config["builder"](config["lr"])
        model.compile(optimizer=Adam(learning_rate=config["lr"]), loss='mse')

        # Train the model
        history = model.fit(X_train, y_train, epochs=config["epochs"], batch_size=32, verbose=0)

        # Evaluate on test set
        y_pred = model.predict(X_test).flatten()
        rmse, mae, r2, mape = compute_metrics(y_test.flatten(), y_pred)
        perf_metrics[config_name] = {"RMSE": rmse, "MAE": mae, "R2": r2, "MAPE": mape}

        # (Optional) Plot predicted vs. actual
        plt.figure(figsize=(6,4))
        plt.scatter(y_test, y_pred, alpha=0.6, label=config_name)
        plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', label="Perfect Fit")
        plt.xlabel("Actual Emissions_Load")
        plt.ylabel("Predicted Emissions_Load")
        plt.title(f"{source} - {config_name}")
        plt.legend()
        plt.grid(True)
        plt.show()

    # Build a summary DataFrame for this powerplant
    summary_data = []
    for config_name, metrics in perf_metrics.items():
        summary_data.append({
            "Configuration": config_name,
            "RMSE": metrics["RMSE"],
            "MAE": metrics["MAE"],
            "R2": metrics["R2"],
            "MAPE": metrics["MAPE"]
        })
    summary_df = pd.DataFrame(summary_data)
    source_summaries[source] = summary_df.copy()

    print(f"\nSummary for {source}:")
    print(summary_df)

    # For each metric, sort and display the best and worst configurations
    metrics_to_sort = {
        "RMSE": "min",
        "MAE": "min",
        "MAPE": "min",
        "R2": "max"  # higher is better
    }
    for metric, order in metrics_to_sort.items():
        ascending = True if order == "min" else False
        sorted_df = summary_df.sort_values(by=metric, ascending=ascending)
        best_config = sorted_df.iloc[0]["Configuration"]
        worst_config = sorted_df.iloc[-1]["Configuration"]
        print(f"\nFor {source} sorted by {metric}:")
        print(sorted_df[["Configuration", metric]])
        print(f"Best {metric}: {best_config} | Worst {metric}: {worst_config}")

# -------------------------------
# (Optional) Overall Summary Across Powerplants
# -------------------------------
for source, df in source_summaries.items():
    print(f"\n=== Overall Summary for {source} ===")
    print(df.sort_values(by="RMSE"))
```
